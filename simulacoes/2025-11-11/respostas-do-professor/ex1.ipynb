```python
import random as rd
def gerar_dados(d,r):
  return [rd.randint(1,r) for n in range(d)]
```

```python
def testar_dados(lista,soma):
  if sum(lista)==soma:
    return True
  return False
```

```python
def simples(d,r,s,l):
  cont = 0
  for k in range(l):
    lista = gerar_dados(d,r)
    if testar_dados(lista,s):
      cont += 1
  return cont
```

```python
d = 1
r = 2
s = 1
l = 100
simples(d,r,s,l)
completa(d,r,s,l,100)
```

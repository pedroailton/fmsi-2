```python
import random as rd
def gerar_dados():
  return [rd.uniform(-1,1) for n in range(2)]
```

```python
gerar_dados()
```

```python
def testar_dados(ponto,raio):
  if ponto[0]**2+ponto[1]**2 <= raio**2:
    return True
  return False
```

```python
# A_c = (P_d)*A_q/P
def simples(raio,numpontos):
  cont = 0
  for k in range(numpontos):
    ponto = gerar_dados()
    if testar_dados(ponto,raio):
      cont += 1
  return cont*4/numpontos
```

```python
raio = 1
numpontos = 100
simples(raio,numpontos)
```

```python
def completa(raio,numpontos,sim):
  resultados = [simples(raio,numpontos) for n in range(sim)]
  return sum(resultados)/sim
```

```python
raio = 1
numpontos = 200
sim = 1000
completa(raio,numpontos,sim)
```
